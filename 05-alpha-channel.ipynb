```python
import cv2
import matplotlib.pyplot as plt
```

Example to read PNG image containing four channels (Red, Green, Blue, Alpha).

```python
img = cv2.imread("./images/transparency.png", cv2.IMREAD_UNCHANGED)
img.shape
```

```python
bgr_image = img[...,0:3]
alpha_image = img[...,-1]
rgb_image = cv2.cvtColor(bgr_image, cv2.COLOR_BGR2RGB)
plt.imshow(rgb_image); plt.title("RGB");
```

```python
plt.imshow(alpha_image, cmap='gray'); plt.title("Alpha");
```

Try with another image

```python
img = cv2.imread("./images/alpha.png", cv2.IMREAD_UNCHANGED)
img.shape
```


<div style="direction:rtl;text-align:right;font-family:Tahoma">
مقدار $\alpha$ مقداری بین 0 تا 255 است که تقسیم بر 255 خواهد شد و فرمول زیر را خواهیم داشت: 
<br/>
پس خواهیم داشت: $0 \leq \alpha \leq 1$</div>

$I = \alpha F + (1 - \alpha) B$
